```python
import pandas as pd
```

```python
import io
import requests
```



# EXCERCISE 1


```python
#assign each url to a variable name

info_url = "https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons.csv"
email_add_url = "https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons_email.csv"
sub_status_url = "https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons_email_chapter_subscription.csv"
```

```python
#read csv from a string 

s = requests.get(info_url).content
df_info = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
#constituent information dataframe

df_info.head()
```

```python
#display all columns

pd.set_option('display.max_columns', None) 
```

```python
#convert to datetime object

df_info['modified_dt'] = pd.to_datetime(df_info['modified_dt'], format = "%a, %Y-%m-%d %X") 
df_info['create_dt'] = pd.to_datetime(df_info['create_dt'], format = "%a, %Y-%m-%d %X")
```

```python
#updated constituent information dataframe

df_info.head()
```

```python
#constituent information schema

df_info.info()
```

```python
#read csv from a string 

s = requests.get(email_add_url).content
df_ea = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
#constituent eamail_address dataframe

df_ea.head()
```

```python
#read csv from a string 

s = requests.get(sub_status_url).content
df_ss = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
#constituent subscription status dataframe

df_ss.head()
```

```python
#convert to boolean data type

df_ss['isunsub'] = df_ss['isunsub'].astype(bool)
df_ss = df_ss[df_ss.chapter_id == 1]
```

```python
#merge the three dataframes

df_merged = df_info.merge(df_ea,on='cons_id').merge(df_ss,on='cons_email_id')
```

```python
#select the given columns 

df_ppl = df_merged[['email', 'subsource', 'isunsub', 'create_dt_x', 'modified_dt_x']]
```

```python
#rename columns

df_ppl.rename(columns = {'subsource':'code', 'isunsub':'is_unsub', 
                         'create_dt_x':'created_dt', 'modified_dt_x':'updated_dt'},
              inplace = True)
```

```python
#display people dataframe

df_ppl.head()
```

```python
#people schema

df_ppl.info()
```

```python
#covert people dataframe to csv file

df_ppl.to_csv("people.csv")
people_file = pd.read_csv("people.csv")
```

```python
print(people_file)
```



# EXCERCISE 2


```python
date_series = df_ppl['created_dt'].dt.date #convert created_dt column from datetime object to date object

date_series.to_frame() #convert series to dataframe
```

```python
#create 2 column dataframe displaying number of constituents acquired on acquisition date

acq = date_series.to_frame().groupby(['created_dt']).size().to_frame()
```

```python
#rename columns

acq.index.name = 'acquisition_date'
acq.rename(columns={0 :'acquisitions'}, inplace=True )
```

```python
acq.head()
```

```python
#acquisition facts schema

acq.info()
```

```python
#convert acuisition facts dataframe to csv file

acq.to_csv("acquisition_facts.csv")
acquisition_facts = pd.read_csv("acquisition_facts.csv")
```
